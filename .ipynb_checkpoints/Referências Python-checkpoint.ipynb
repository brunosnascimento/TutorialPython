```python
# Definindo uma função
def f(x):
    return np.exp(-x ** 2)
```

```python
x = np.linspace(-1, 5, num=1000)
x
```

```python
plt.plot(x, f(x))
plt.xlabel("Eixo x")
plt.ylabel("f(x)")
plt.legend("Teste de legenda")
```

```python
# Plotando com grid
plt.plot(x, f(x), label="Função f(x)")
plt.xlabel("Eixo x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend() # criando a legenda a partir do label definido
```

```python
# Modificando cores e o tipo de linha
plt.plot(x, f(x), 'og', label="Função f(x)")
plt.xlabel("Eixo x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend() # criando a legenda a partir do label definido
```

```python
# Mais exemplos
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

<h4> Integração numérica </h4>

```python
# Importando um pacote específico
from scipy import integrate
```

```python
%%latex
$$f(x) = e^{-x^2}$$
```

```python
def f(x):
    return np.exp(-x ** 2)
```

```python
integrate.quad(f, 0, 5) # output valor da integração, extimativa de erro
```

```python
%%latex
# Adicionando parâmetros extras (constante A)
$$f(x) = Ae^{-x^2}$$ 
```

```python
def f(x, A):
    return A * np.exp(-x ** 2)
```

```python
integrate.quad(f, 0, 5, args=1.0) # parâmetro args permite passa argumentos para f(x)
```

<h4> Equações diferenciais ordinárias EDO</h4>

```python
%%latex
$$y' + y = 0$$
$$y' = f(y)$$
```

```python
def f(y, t):
    return -y
```

```python
y0 = 1.0
t = np.linspace(0,3)
```

```python
sol = integrate.odeint(f, y0, t)
```

```python
plt.plot(t, sol)
```

<h3> Equações algebricas não lineares </h3>

```python
from scipy import optimize
```

```python
%%latex
$$f(E) = E - esinE - M$$
```

```python
def F(E, e, M):
    return E - e * np.sin(E) - M
```

```python
optimize.root(F, 0.1, args=(0.016, 0.1))
```

<h3> Interpolação </h3>

https://www.pybonacci.org/tag/interpolacion/

```python
# importando o pacote
from scipy.interpolate import barycentric_interpolate
```

```python
def runge(x):
    '''Função de Runge'''
    return 1 / (1 + x ** 2)
```

```python
N = 11 # pontos de interpolação
xp = np.arange(11)
xp
```

```python
fp = runge(xp)

x = np.linspace(-5, 5, num=1000)

y = barycentric_interpolate(xp, fp, x)
```

<h2> Parte 6 - Widgets interativos</h2>

```python
from IPython.html.widgets import interact
```

```python
def interact_function(freq):
    x = np.linspace(0, 3, num=200)
    y = x * np.sin(freq * x)
    plt.plot(x, y, color='red')
```

```python
interact(interact_function(np.pi), freq=(2 * np.pi, 2 * np.pi * 10.0))
```
